```python
df = df[['Country', 'Area(sq km)', 'Birth rate(births/1000 population)', 'Death rate(deaths/1000 population)',
           'Debt - external', 'Electricity - consumption(kWh)',
       'Electricity - production(kWh)', 'Exports', 'GDP','Highways(km)', 'Imports','Inflation rate (consumer prices)(%)',
       'Internet users','Labor force',
       'Life expectancy at birth(years)',
       'Military expenditures - dollar figure', 'Oil - consumption(bbl/day)', 
       'Population','Total fertility rate(children born/woman)', 'Unemployment rate(%)']]
```

### Importing seaborn library

```python
import seaborn as sns
```

### Plotting a boxplot to identify outliers in the dataset

```python
sns.boxplot('Exports',data=df)
```

### Removing outliers

```python
df = df[df['Exports']<0.75e11]
```

### Checking for outliers in other attributes

```python
sns.boxplot('Debt - external',data=df)
```

### Plotting a displot to get the distribution in the form of a histogram. By default Kernel Density Estimation is also provided.

```python
sns.set_style('whitegrid')
sns.distplot(df['Electricity - consumption(kWh)'], color ='green', bins = 10)
```

### Histogram is plotted after turing off the KDE. We see that the Electricity Consumption is concentrated between 0.1 to 0.5e11 KWh

```python
sns.distplot(df['Electricity - consumption(kWh)'], kde = False, color ='green', bins = 10)
```

### Plotting a Histogram for GDP. We got to know that most of the countries have GDP below 0.5e12

```python
sns.distplot(df['GDP'], kde = False, color ='blue', bins = 25)
```

### Importing matplotlib library

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### Plotting the scatter plot of Exports and Area of countries. We see that in small to medium sized countries, Area matters for amount in exports from that country but that does not hold true for medium to large sized countries

```python
sns.jointplot(x = df["Exports"], y = df["Area(sq km)"], kind = "scatter")
```

### Here we tried to fit a regression model on the scatter plot and it verifies our observation as the line fails to cover the points after a certain area and exports

```python
sns.jointplot(x = df["Exports"], y = df["Area(sq km)"], kind = "reg")
```

### This is a KDE plot between Life Expectancy and Population. We can see that with increasing population the density goes on the lower side whereas for countries with low population or medium ones, the expectancy reaches around 70 years

```python
sns.jointplot(x = df["Life expectancy at birth(years)"], y = df["Population"], kind = "kde")
```

### This plot shows that the labor force does not decide the unemployment rate. It shows unemployment increases irrespective of the labor force size.

```python
sns.jointplot(x = df["Unemployment rate(%)"], y = df["Labor force"], kind = "scatter")
```
